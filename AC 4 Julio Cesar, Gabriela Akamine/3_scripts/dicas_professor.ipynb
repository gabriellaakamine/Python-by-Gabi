```python
import pandas as pd

dados = pd.read_csv('../1_bases_tratadas/base.csv')
dados.rename(columns={'quant_celular': 'Quantidade de celulares registrados por BO'}, inplace=True)
dados.rename(columns={'marca_celular': 'Marcas dos celulares'}, inplace=True)
```

```python
dados
```

```python
auxiliar = dados.groupby(['peridoocorrencia'])['peridoocorrencia'].agg(['count'])
auxiliar
```

```python
auxiliar= auxiliar.reset_index().rename(columns={'count':'qtdperiodo'})
auxiliar
```

```python
dados = pd.merge(dados, auxiliar, how='left', on='peridoocorrencia')
dados
```

```python
# replicar os passos acima para a variável cidade
auxiliar2 = dados.groupby(['cidade'])['cidade'].agg(['count'])
auxiliar2
```

```python
auxiliar2 = auxiliar2.reset_index().rename(columns={'count':'qtdcidades'})
auxiliar2
```

```python
auxiliar3 = dados.groupby(['solucao'])['solucao'].agg(['count'])
auxiliar3
```

```python
auxiliar3 = auxiliar3.reset_index().rename(columns={'count':'qtdsolucao'})
```

```python
dados = pd.merge(dados, auxiliar2, how='left', on='cidade')
dados
```

```python
dados = pd.merge(dados, auxiliar3, how='left', on='solucao')
dados
```
